```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
plt.style.use(['science', 'notebook', 'grid'])
```

```python
#Parameter
A = 0.1    #m
l = 1      #m
Wd = 100   #s^-1
g = 9.8    #m/s

#Array handling
startTime = 0
endTime = 5
N = 1000000  #Increse N as much as your computer can 
t, dt = np.linspace(startTime, endTime, N, retstep=True)
tht = np.zeros((N), )
w = np.zeros((N), )

#Initial condition
tht[0] = np.pi/3
w[0] = 0

#Euler Method
for i in range(N-1):
    tht[i+1] = tht[i] + dt*w[i]
    w[i+1] = w[i] + dt*(g - A*Wd**2*np.cos(Wd*t[i])*(tht[i]/l))
    
#Plotting
plt.figure(figsize = (18, 6))
plt.title("Inverted pendulum(Simplified)")
plt.plot(t, (180/np.pi)*tht)
plt.xlabel(r"$t$", size = 20)
plt.ylabel(r"$\theta$(degree)", size = 20)
plt.show()
```

```python
x = l*np.sin(tht)
y = l*np.cos(tht) + A*np.cos(w*t)

plt.figure(figsize = (18, 6))
plt.plot(tht, w)
```
